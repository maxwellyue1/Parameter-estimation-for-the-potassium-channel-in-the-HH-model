```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def steady_state(u, V_2n, s_n): 
    return 1 / (1 + np.exp((u - V_2n) / s_n))
```

```python
def time_constant(u, tau_0n, delta_n, V_2n, s_n):
    return (tau_0n * np.exp(delta_n*((u-V_2n)/s_n))) / (1+np.exp((u-V_2n)/s_n))
```

```python
u = np.linspace(-100, 100, 1000)
V_2n = 0
s_n = -15
```

```python
plt.plot(u, steady_state(u, V_2n, s_n), color='black', linestyle='solid', label = '$V_{2n}$')
plt.plot(u, steady_state(u, V_2n-20, s_n), color='black', linestyle='dashed', label = '$V_{2n}-20$')
plt.plot(u, steady_state(u, V_2n+20, s_n), color='black', linestyle='dotted', label = '$V_{2n}+20$')

plt.xticks([V_2n-20, V_2n, V_2n+20], ['$V_{2n}-20$', '$V_{2n}$', '$V_{2n}+20$'])
plt.yticks([0, 0.5, 1])

plt.xlabel('u (mV)')
plt.ylabel('$n_\infty$')
plt.legend()

plt.axvline(x=V_2n, ymin=0, ymax=0.5, color='red', linestyle='--', alpha=0.3)
plt.axvline(x=V_2n-20, ymin=0, ymax=0.5, color='red', linestyle='--', alpha=0.3)
plt.axvline(x=V_2n+20, ymin=0, ymax=0.5, color='red', linestyle='--', alpha=0.3)
plt.plot([-100, V_2n+20], [0.5, 0.5], color='red', linestyle='--', alpha=0.3)


plt.margins(0.001)
plt.show()

```

```python
# u = np.linspace(-140, 140, 1000)

plt.plot(u, steady_state(u, V_2n, s_n), color='black', linestyle='solid', label = '$s_n$')
plt.plot(u, steady_state(u, V_2n, 2 * s_n), color='black', linestyle='dashed', label = '$2s_n$')
plt.plot(u, steady_state(u, V_2n, 0.5 * s_n), color='black', linestyle='dotted', label = '$s_n/2$')

plt.xticks([V_2n], ['$V_{2n}$'])
plt.yticks([0, 0.5, 1])

plt.axvline(x=V_2n, ymin=0, ymax=0.5, color='red', linestyle='--', alpha=0.3)
plt.plot([-100, V_2n], [0.5, 0.5], color='red', linestyle='--', alpha=0.3)

plt.xlabel('u (mV)')
plt.ylabel('$n_\infty$')
plt.legend()

plt.margins(0.001)
plt.show()
```

```python
u = np.linspace(-100, 100, 1000)
tau_0n = 20
delta_n = 0.5
V_2n = 0
s_n = -15
```

```python
plt.plot(u, time_constant(u, tau_0n, delta_n, V_2n, s_n), color='black', linestyle='solid', label = '$V_{2n}$')
plt.plot(u, time_constant(u, tau_0n, delta_n, V_2n-20, s_n), color='black', linestyle='dashed', label = '$V_{2n}+20$')
plt.plot(u, time_constant(u, tau_0n, delta_n, V_2n+20, s_n), color='black', linestyle='dotted', label = '$V_{2n}-20$')

plt.xticks([V_2n-20, V_2n, V_2n+20], ['$V_{2n}-20$', '$V_{2n}$', '$V_{2n}+20$'])
plt.yticks([])

plt.xlabel('u (mV)')
plt.ylabel('$n_\infty$')
plt.legend()

plt.axvline(x=V_2n, color='red', linestyle='--', alpha=0.3)
plt.axvline(x=V_2n-20, color='red', linestyle='--', alpha=0.3)
plt.axvline(x=V_2n+20, color='red', linestyle='--', alpha=0.3)
# plt.plot([-100, V_2n+20], [0.5, 0.5], color='red', linestyle='--', alpha=0.3)


plt.margins(0.001)
plt.show()
```

```python
plt.plot(u, time_constant(u, tau_0n, delta_n, V_2n, s_n), color='black', linestyle='solid', label = '$\\tau_{0n}$')
plt.plot(u, time_constant(u, tau_0n*2, delta_n, V_2n, s_n), color='black', linestyle='dashed', label = '$2 \\tau_{0n}$')
plt.plot(u, time_constant(u, tau_0n/2, delta_n, V_2n, s_n), color='black', linestyle='dotted', label = '$\\tau_{0n}/2$')

plt.xticks([])
plt.yticks([])

plt.xlabel('u (mV)')
plt.ylabel('$\\tau_n$')
plt.legend()

plt.margins(0.001)
plt.show()
```

```python
plt.plot(u, time_constant(u, tau_0n, 1, V_2n, s_n), color='black', linestyle='solid', label = '$1$')
plt.plot(u, time_constant(u, tau_0n, 0.25, V_2n, s_n), color='black', linestyle='--', label = '$0.25$')
plt.plot(u, time_constant(u, tau_0n, 0.5, V_2n, s_n), color='black', linestyle=':', label = '$0.5$')
plt.plot(u, time_constant(u, tau_0n, 0.75, V_2n, s_n), color='black', linestyle='-.', label = '$0.5$')
plt.plot(u, time_constant(u, tau_0n, 0, V_2n, s_n), color='black', linestyle='solid', label = '$0$')

plt.text(-90, 19, '0', fontsize=9, color='red')
plt.text(90, 19, '$1$', fontsize=9, color='red')

plt.text(-70, 8, '0.75', fontsize=9, color='red')
plt.text(60, 8, '$0.25$', fontsize=9, color='red')

plt.text(30, 7, '$0.5$', fontsize=9, color='red')

plt.xticks([])
plt.yticks([])

plt.xlabel('u (mV)')
plt.ylabel('$\\tau_n$')
# plt.legend()

plt.margins(0.001)
plt.show()
```
